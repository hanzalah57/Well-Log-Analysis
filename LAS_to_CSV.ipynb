```python
import lasio
import pandas as pd

def las_to_csv_custom(las_file, csv_file):
    """
    Convert LAS file to CSV.
    
    Parameters:
    -----------
    las_file : str
        Path to input LAS file.
    csv_file : str
        Path to output CSV file.
    """
    # Read LAS file
    las = lasio.read(las_file)
    
    # Convert to DataFrame
    df = las.df()
    
    # Bring depth into first column (reset_index adds DEPT as a column)
    df = df.reset_index()
    df.rename(columns={"DEPT": "DEPT"}, inplace=True)
    
    # Save to CSV
    df.to_csv(csv_file, index=False)
    print(f"CSV file written to: {csv_file}")
```

```python
las_to_csv_custom("FALKON-1.las", "FALKON-1.csv")
```
