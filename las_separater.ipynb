```python
def split_multi_version_las(input_file):
    #This function splits different versions prtesent in singale well las file
    import os

    # Read entire file
    with open(input_file, 'r', encoding='utf-8', errors='ignore') as file:
        lines = file.readlines()

    # Identify version start indices
    version_indices = [i for i, line in enumerate(lines) if "Version Information" in line or "~Version" in line]

    if len(version_indices) <= 1:
        print("Only one version found. No split needed.")
        return

    base_name = os.path.splitext(os.path.basename(input_file))[0]
    output_dir = os.path.dirname(input_file)

    for idx, start in enumerate(version_indices):
        end = version_indices[idx + 1] if idx + 1 < len(version_indices) else len(lines)
        version_lines = lines[start:end]

        version_filename = os.path.join(output_dir, f"{base_name}_version_{idx + 1}.las")
        with open(version_filename, 'w') as version_file:
            version_file.writelines(version_lines)
        print(f"Created: {version_filename}")
```
